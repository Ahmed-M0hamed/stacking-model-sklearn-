```python
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestClassifier , GradientBoostingClassifier 
from sklearn.linear_model import LogisticRegression
```

```python
class stacking : 
  def __init__(self) : 
    self.rf = RandomForestClassifier()
    self.gb = GradientBoostingClassifier()
    self.lr = LogisticRegression(fit_intercept=False)
  def fit(self ,train_data , train_labels ): 
    features = train_data.columns 
    part_1 , part_2 , part_1_labels , part_2_labels = train_test_split(train_data , train_labels, test_size=0.5, random_state=123)
    self.rf.fit(part_1[features] , part_1_labels)
    self.gb.fit(part_1[features],part_1_labels )
    part_2['rf_pred'] = self.rf.predict(part_2[features])
    part_2['gb_pred'] = self.gb.predict(part_2[features])
    self.lr.fit(part_2[['gb_pred', 'rf_pred']],part_2_labels )
  def predict(self , test_data):
    features = test_data.columns
    test_data['gb_pred'] = self.gb.predict(test_data[features])
    test_data['rf_pred'] = self.rf.predict(test_data[features])
    preds = self.lr.predict(test_data[['gb_pred', 'rf_pred']])
    return preds 
```
